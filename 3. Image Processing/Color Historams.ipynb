```python
# Histgram is a visual representation of distirbution of a continuous feature  
```

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
dark_horse = cv2.imread('/home/linux/Desktop/Computer Vision/Image Processing/DATA/horse.jpg')# ORIGINAL BGR OPENCV
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB) # CONVERTED TO RGB FOR SHOW

rainbow = cv2.imread('/home/linux/Desktop/Computer Vision/Image Processing/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('/home/linux/Desktop/Computer Vision/Image Processing/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_bricks)
```

```python
# OPENCV BGR 
hist_values = cv2.calcHist([blue_bricks], channels =[0], mask = None, histSize = [256], ranges = [0,256])
```

```python
plt.plot(hist_values) # histgram value of blue channel in bricks image 
```

```python
hist_values = cv2.calcHist([dark_horse], channels =[0], mask = None, histSize = [256], ranges = [0,256])
plt.plot(hist_values) # histgram value of blue channel in horse image 
# we note that blue level is close to zero because image has little blue color on it
```

```python
# SHOW HISTGRAM FOR B & G & R FOR BRICKS
img = blue_bricks

color = ('b', 'g', 'r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256]) # limits of x
    
plt.title('HISTGRAM FOR BLUE BRICKS')
```

```python
# SHOW HISTGRAM FOR B & G & R FOR HORSE
img = dark_horse

color = ('b', 'g', 'r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,50]) # limits of x
    plt.ylim([0,500000]) # change limits to make it closer
    
plt.title('HISTGRAM FOR BLUE HORSE')
```

```python
gorilla = cv2.imread('/home/linux/Desktop/Computer Vision/Image Processing/DATA/gorilla.jpg', 0)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(gorilla)
```

```python
hist_values = cv2.calcHist([gorilla], channels = [0], mask = None, histSize = [256], ranges = [0,256] )
```

```python
plt.plot(hist_values)
```

```python
# Equaliztion Histgram
eq_gorilla = cv2.equalizeHist(gorilla)
# increase high contrast of image
```

```python
display_img(eq_gorilla)
```

```python
hist_values = cv2.calcHist([eq_gorilla], channels = [0], mask = None, histSize = [256], ranges = [0,256] )
```

```python
plt.plot(hist_values)
```
