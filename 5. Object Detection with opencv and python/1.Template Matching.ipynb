```python
full = cv2.imread('/home/linux/Desktop/Computer Vision/5. Object Detection with opencv and python/DATA/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
face = cv2.imread('/home/linux/Desktop/Computer Vision/5. Object Detection with opencv and python/DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
# we want to get match of dog face in larger image 
# by sanc each pixel on sammy image to find match face
```

```python
# Eval fuction its used to get match of things
# we can convert string 'sum' to sum fuction by using eval

mystring = 'sum'
myfunc = eval(mystring) # this means it found 'sum' match sum function
myfunc([1,2,3])
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings ,later on we 'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED' ]
```

```python
 for m in methods:
    
    # CREATE A COPY OF IMAGE TO DRAW ON IT
    full_copy = full.copy()
    
    method = eval(m)
    
    # TEMPLATE MATCHING 
    res = cv2.matchTemplate(full_copy, face, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    # to check topleft of result 
    if method in [cv2.TM_SQDIFF ,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc  # because cv2.TM_SQDIFF ,cv2.TM_SQDIFF_NORMED this two mehods is took minloc as topleft
    else:
        top_left = max_loc  # because other methods are took maxloc as topleft
        
    
    hieght,width,channels = face.shape
    
    bottom_right = (top_left[0]+width, top_left[1]+hieght)
    
    cv2.rectangle(full_copy, top_left, bottom_right, (255,0,0), thickness =10)
    
    # PLOT AND SHOW IMAGE
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEATMAP OF TEMPLATE MATCHING')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('DETECTION OF TEMPLATE')
    
    #TITLE WITH THE METHOD USED
    plt.title(m)
    
    plt.show()
    
    print('/n')
    print('/n')
```
